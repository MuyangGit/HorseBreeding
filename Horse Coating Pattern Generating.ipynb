```python
import random

from matplotlib import pyplot as plt;
import numpy as np;
```

```python
def getCoat(size,percentage, soft_edge):

    np.random.seed(20)

    plt.figure(figsize=(30,10))
    plt.subplot(131);

    m,n = 200,200;
    base_pattern = np.random.randint(225, size=(m,n, 1))

    pattern_1 = [];
    radius = 10;
    percentage_threshold = int(percentage/100*225);
    print(percentage_threshold)
    for i in range(m):
        row = [];
        for j in range(n):
            area = base_pattern[i-radius:i+radius,j-radius:j+radius];
            row.append(percentage_threshold if np.mean(area) >= percentage_threshold else 0)
        pattern_1.append(row);
    pattern_1 = np.array(pattern_1)
    plt.subplot(131)
    plt.title("Horse Body Coating")
    plt.imshow(pattern_1,cmap = "Greys")

    pattern_2 = [];
    radius = 30;
    size_threshold = int( size / 100 * percentage_threshold )
    for i in range(m):
        row = [];
        for j in range(n):
            area = pattern_1[i-radius:i+radius,j-radius:j+radius];
            row.append(np.mean(area)<size_threshold);
        pattern_2.append(row);
    pattern_2 = np.array(pattern_2)

    plt.subplot(132)
    plt.title("Horse Body Coating")
    plt.imshow(pattern_2,cmap = "Greys")

    pattern_3 = [];
    radius = 20;
    for i in range(m):
        row = [];
        for j in range(n):
            area = pattern_2[i:i+soft_edge,j:j+soft_edge];
            row.append(np.mean(area));
        pattern_3.append(row);
    pattern_3 = np.array(pattern_3)

    plt.subplot(133)
    plt.title("Horse Body Coating")
    plt.imshow(pattern_3,cmap = "Greys")
    plt.show()

    return pattern_2

getCoat(20, 50.5, 10);
```
